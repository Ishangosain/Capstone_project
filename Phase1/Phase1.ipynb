## Population Data Collection Refrence:

Population in India state wise data:    
https://en.wikipedia.org/wiki/List_of_states_in_India_by_past_population

With above data I got the population of each state for the year 2001 and the year 2011. For getting the data from 2002-2010,I calculated average population growth using this formula:- minus the data of 2011 with 2001 data and then divide it with 11 years. Using this I got the average population from 2002-2010,so I made full daat from 2001-2011.

```python
population_India=pd.read_csv('population_India.csv')
```

```python
population_India.head()
```

```python
population_India = population_India.rename(columns={'2001':'Year2001','2002':'Year2002','2003':'Year2003','2004':'Year2004',
                       '2005':'Year2005','2006':'Year2006','2007':'Year2007','2008':'Year2008','2009':'Year2009','2010':'Year2010',
                      '2011':'Year2011','2012':'Year2012'})
```

```python
population_India=population_India.drop(columns='Unnamed: 13')
```

```python
population_India.head()
```

```python
population_India = pd.melt(population_India, id_vars=['States & UT'], var_name='Year', value_name='population')
```

```python
population_India
```
